This notebook performs comprehensive exploratory data analysis on the Crop Recommendation dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

print('Libraries imported successfully')
```

```python
df = pd.read_csv('../dataset/Crop_recommendation.csv')
print(f'Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns')
```

## 4. Crop Distribution

```python
plt.figure(figsize=(12, 6))
counts = df['label'].value_counts()
colors = sns.color_palette('Set2', len(counts))
bars = plt.bar(counts.index, counts.values, color=colors)
plt.title('Crop Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Crop', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1,
             str(int(bar.get_height())), ha='center', fontsize=10)
plt.tight_layout()
plt.show()
print(f'Unique crops: {df["label"].nunique()}')
```

## 5. Numerical Feature Distributions

```python
fig, axes = plt.subplots(2, 4, figsize=(16, 10))
features = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
colors = ['#4CAF50', '#FF9800', '#2196F3', '#FF5722', '#9C27B0', '#4CAF50', '#E91E63']

for i, (feature, color) in enumerate(zip(features, colors)):
    row, col = i // 4, i % 4
    axes[row, col].hist(df[feature], bins=20, color=color, edgecolor='black', alpha=0.7)
    axes[row, col].set_title(f'{feature} Distribution', fontsize=12, fontweight='bold')
    axes[row, col].set_xlabel(feature)
    axes[row, col].set_ylabel('Frequency')
    axes[row, col].axvline(df[feature].mean(), color='red', linestyle='--', label=f'Mean: {df[feature].mean():.2f}')
    axes[row, col].legend(fontsize=8)

fig.delaxes(axes[1, 3])
plt.tight_layout()
plt.show()
```

## 6. Boxplots by Crop

```python
fig, axes = plt.subplots(3, 3, figsize=(16, 14))
axes = axes.flatten()

for i, col in enumerate(features):
    df.boxplot(column=col, by='label', ax=axes[i])
    axes[i].set_title(f'{col} by Crop', fontsize=11, fontweight='bold')
    axes[i].set_xlabel('')
    axes[i].tick_params(axis='x', rotation=45, labelsize=8)

for j in range(len(features), len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('')
plt.tight_layout()
plt.show()
```

## 7. Correlation Matrix and Heatmap

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='RdYlGn',
            square=True, linewidths=0.5, cbar_kws={'shrink': 0.8})
plt.title('Correlation Matrix Heatmap', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print('\nCorrelation values:')
print(corr)
```

## 8. Pairplot

```python
sample = df.sample(min(500, len(df)), random_state=42)
g = sns.pairplot(sample[numeric_cols], diag_kind='kde', plot_kws={'alpha': 0.5})
g.fig.suptitle('Pairplot of Numerical Features', y=1.02, fontsize=16, fontweight='bold')
plt.show()
```

## 9. Nutrient Analysis (NPK)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
nutrients = ['N', 'P', 'K']
colors = ['#4CAF50', '#FF9800', '#2196F3']

for i, (nutrient, color) in enumerate(zip(nutrients, colors)):
    axes[i].hist(df[nutrient], bins=20, color=color, edgecolor='black', alpha=0.7)
    axes[i].set_title(f'{nutrient} Distribution', fontsize=14, fontweight='bold')
    axes[i].set_xlabel(nutrient)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
print('Average nutrient levels by crop:')
print(df.groupby('label')[['N', 'P', 'K']].mean().round(1))
```

## 10. Environmental Analysis

```python
print('\nAverage environmental conditions by crop:')
print(df.groupby('label')[['temperature', 'humidity', 'ph', 'rainfall']].mean().round(1))
```
